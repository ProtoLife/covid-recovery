# Map of consensus clustering from Cluster notebook

```python
import folium
import pandas as pd
```

```python
import pickle as pk
```

```python
#Setting up the world countries data URL
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country_shapes = f'{url}/world-countries.json'
```

Read data from Cluster notebook

```python
with open('clusalign.pk','rb') as fp:
    clusalign = pk.load(fp)
    
clusalign;
```

```python
df = []
for cc in clusalign:
    for country in clusalign[cc]:
        if cc == -1:
            df.append([country,1])
        if cc == 0:
            df.append([country,2])
        if cc == 1:
            df.append([country,3])
        if cc == 2:
            df.append([country,4])
        if cc == 3:
            df.append([country,5])
        if cc == 4:
            df.append([country,6])
df_clusalign = pd.DataFrame(df, columns = ['Country', 'Total Case'])
```

```python
df_clusalign.replace('United States', 'United States of America', inplace=True)
#Replacing the country name
df_clusalign.replace('USA', "United States of America", inplace = True)
df_clusalign.replace('Tanzania', "United Republic of Tanzania", inplace = True)
df_clusalign.replace('Democratic Republic of Congo', "Democratic Republic of the Congo", inplace = True)
df_clusalign.replace('Congo', "Republic of the Congo", inplace = True)
df_clusalign.replace('Lao', "Laos", inplace = True)
df_clusalign.replace('Syrian Arab Republic', "Syria", inplace = True)
df_clusalign.replace('Serbia', "Republic of Serbia", inplace = True)
df_clusalign.replace('Czechia', "Czech Republic", inplace = True)
df_clusalign.replace('UAE', "United Arab Emirates", inplace = True)
```

```python
m = folium.Map(location=[5., 12.],zoom_start=2)
#Adding the Choropleth layer onto our base map
folium.Choropleth(
    #The GeoJSON data to represent the world country
    geo_data=country_shapes,
    name='choropleth COVID-19',
    data=df_clusalign,
    #The column aceppting list with 2 value; The country name and  the numerical value
    columns=['Country', 'Total Case'],
    key_on='feature.properties.name',
    # fill_color='YlGn',
    fill_color='Spectral',
    nan_fill_color='white'
).add_to(m)
m
```
